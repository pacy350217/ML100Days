### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = "utf-8"
response = r.text
print(response)
print(type(response)) #<class 'str'>


soup = BeautifulSoup(response,"html5lib")
print(soup)
print(type(soup)) #<class 'bs4.BeautifulSoup'>
print(soup.html["lang"]) #zh-TW
print(soup.head["itemtype"]) #"https://schema.org/WebSite"
print(soup.meta["charset"]) #utf-8
for tag in soup.find_all("meta"):
    print(tag)

```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

import requests

url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = "utf-8"
response = r.text
print(response)
print("資料型態為：",type(response)) #<class 'str'>
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

import requests
from bs4 import BeautifulSoup


soup = BeautifulSoup(response,"html5lib")
print(soup)
print("BeautifulSoup可以解析HTML文件，資料型態為：",type(soup)) #<class 'bs4.BeautifulSoup'>
print(soup.html["lang"]) #zh-TW
print(soup.head["itemtype"]) #"https://schema.org/WebSite"
print(soup.meta["charset"]) #utf-8
for tag in soup.find_all("meta"):
    print(tag)

```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
headers = {"user-agent":"my-app/0.0.1"}
r = requests.get(url,headers=headers)
r.encoding = 'utf-8'
response = r.text
print(response)
print(type(response)) #<class 'str'>
print(response[0:600])

soup = BeautifulSoup(response,"html5lib")
print(type(soup))
print(soup.html["data-theme"])
print(soup.html["data-hairline"])
tags = soup.find_all("meta",content="webkit") #[<meta content="webkit" name="renderer"/>, <meta content="webkit" name="force-rendering"/>]
print(tags)
print("---------------------")
tags_1 = soup.find_all("meta")
print(tags_1)
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
