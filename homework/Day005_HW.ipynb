## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests, json 
r = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
datas = json.loads(response)

#回傳資料筆數
count = 0 
for data in datas:
    data["id"]
    count += 1

print("回傳資料筆數：",count)

#包含欄位
column = []
for data in datas:
    
    for d in data.keys():
        if d not in column:
            column.append(d)

print(column)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

import requests, json 
r = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
datas = json.loads(response)


for data in datas:
    #取出標題
    print(data["title"])
    #取出貼文時間
    print(data["createdAt"])
    #留言人數
    print(data["commentCount"])
    #按讚人數
    print(data["likeCount"])


```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

#平均留言人數
commentCount = []
for data in datas: 
    count1 = data["commentCount"]
    commentCount.append(count1)
    
commentAmount = int(sum(commentCount)/30)
print("平均留言人數：",commentAmount)


#平均按讚人數
likeCount = []
for data in datas: 
    count = data["likeCount"]
    likeCount.append(count)
    
  
likeAmount = int(sum(likeCount)/30)
print("平均按讚人數：",likeAmount)
```
